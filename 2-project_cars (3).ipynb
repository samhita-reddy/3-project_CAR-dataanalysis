### Does higher price of the car guarantee less or more weight?Find out the relation between them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
#numeric vs numeric first
sns.scatterplot(data = car,x = 'MSRP',y = 'Weight',hue='Origin',palette='Dark2') #similar to line nd scatter plots
sns.rugplot(data = car,x = 'MSRP',y = 'Weight',color='brown') # we can also observe the density parallely
#observations:
# ASIAN CARS HAVE LESS WEIGHT AND LESS MSRP 
# MOST CARS IN USA ALMOST HAVE AROUND THE SAME MSRP COMPARED TO EUROPE OR ASIA
```

### What is the relation between the LENGHTS and WEIGHTS of cars?

```python
sns.kdeplot(data=car, x="Length", y="Weight",fill=True)
#the perks of using this over scatter plot/lineplot/regplot? 
#THIS CLEARLY SHOWS THE DENSITY WHERE DARKER AREAS= MOST OF THE CARS,LIGHTER AREAS=FEW CARS IN THAT REGION
# observation:
# HEAVIER CARS HAVE MORE LENGTH.
# MOST CARS: Length â‰ˆ 175â€“190. Weight â‰ˆ 3000â€“3600. 
# MOST CARS IN THIS DATASET ARE MID LENGTH AND MID WEIGHT VEHICLES 
```

```python
sns.relplot(data = car,x = 'MSRP', y = 'Horsepower',col='Origin',kind='scatter',alpha=0.6,s=60,)
# the clusters are clearly visible to differentiate for a general clarity 
# and now the actaul plot is the below cell.
```

### Does higher price guarantee more horsepower?Find out the relation between price of cars and its horsepower

```python
sns.lmplot(
    data=car,
    x="MSRP",
    y="Horsepower",
    hue="Origin",
    scatter_kws={"alpha": 0.5, "s": 50},
    line_kws={"linewidth": 3}
)
# observations :
# USA cars appear more horsepower-oriented at comparable price points.
# High price does NOT guarantee high horsepower for European cars.
# European cars show high price variability with less consistent performance gains,reflecting a stronger focus on luxury and brand value.
```

### Compare the speed of the car in CITY VS HIGHWAY

```python
sns.set_style("dark")
sns.regplot(data=car,x = 'MPG_City',y = 'MPG_Highway',color='turquoise')
sns.despine()
#observations
# Highway MPG is always higher than City MPG,City: 18â€“30 MPG, Highway: 22â€“35 MPG
```

### Compare the PRICE of different cars based on its origin and type

```python
# CATEGORICAL VS NUMERICAL
sns.set_style("white")
sns.boxplot(data=car, x="Type", y="MSRP", hue="Origin")
sns.stripplot(data = car,x = 'Type',y = 'MSRP',hue='Origin',dodge=True,alpha=0.6)
# observations:
# SPORTS cars exhibit the highest price dispersion driven largely by European brands.
# SEDANS form the largest and most diverse segment, while SUVs and TRUCKs show more standardized pricing
```

### What is the average length of different types of cars and which are the longest nd shortest?

### How does the region of origin relate to car manufacturers, and which brands are produced in each region?

```python
table = pd.crosstab(car["Make"], car["Origin"])
sns.heatmap(table, cmap="Reds")

#The heatmap reveals a strong association between car manufacturer and region of origin.
# Most brands are almost entirely confined to a single geographic origin, with minimal overlap across Asia,
# Europe, and the USA. This indicates that Make and Origin are highly dependent categorical features in the dataset.â€
```
